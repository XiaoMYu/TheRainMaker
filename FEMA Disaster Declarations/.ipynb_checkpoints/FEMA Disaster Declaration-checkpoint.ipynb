```python
#Dependencies

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime
```

```python
# Store file path.
file = ("Resources/FEMA_Disaster incidents.xlsx")

xl = pd.ExcelFile("Resources/FEMA_Disaster incidents.xlsx")

print(xl.sheet_names)
```

```python
fema_declarations = xl.parse("FEMA Declarations")
fema_declarations
```

```python
fema_declarations['year'] = fema_declarations['Declaration Date'].dt.year
fema_declarations.head()
```

```python
declaration_year = fema_declarations.groupby(["year"]).count()
declaration_year.reset_index(level=0, inplace=True)
declaration_year
```

```python
year_data = declaration_year[["year", "Incident Type"]]
year_data
```
